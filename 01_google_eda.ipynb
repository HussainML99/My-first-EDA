- Let's convert the `M` and `K` units into bytes and then remove the `M` and `K` from the values and convert them into numeric data type.

---

````python
# convert the size column to numeric by multiplying the values with 1024 if it has 'k' in it and 1024*1024 if it has 'M' in it
# this function will convert the size column to numeric
def convert_size(size):
    # add function details here
    '''
    This function will convert the size column to numeric by multiplying the values with 1024 if it has 'k' in it and 1024*1024 if it has 'M' in it
    '''
    
    if isinstance(size, str):
        if 'k' in size:
            return float(size.replace('k', '')) * 1024
        elif 'M' in size:
            return float(size.replace('M', '')) * 1024 * 1024
        elif 'Varies with device' in size:
            return np.nan
    return size

df['size'] = df['size'].apply(convert_size)
````

- let's change the name `size` to `size_in_bytes`

````python
# rename the column name 'Size' to 'Size_in_bytes'
df.rename(columns={'size': 'size_in_bytes'}, inplace=True)
````

- let's we make a new columns `size_in_mb`

````python
# making a new column called 'Size in Mb' which will have the size in MB
df['size_in_mb'] = df['size_in_bytes'].apply(lambda x: x/(1024*1024))
````

## **Question 19:** 
### **Which genres have the highest and lowest ratings on average, and what could be the reason for this?**

````python
# Assuming the dataset has 'genre' and 'rating' columns
# Split genres if they are in a single string
df['genre'] = df['genres'].str.split(',')

# Explode the dataframe so that each genre has its own row
df_exploded = df.explode('genres')

# Calculate average ratings by genre
average_ratings = df_exploded.groupby('genres')['rating'].mean().reset_index()

# Find genres with highest and lowest average ratings
highest_rated = average_ratings.loc[average_ratings['rating'].idxmax()]
lowest_rated = average_ratings.loc[average_ratings['rating'].idxmin()]

print("Highest Rated Genre:")
print(highest_rated)

print("\nLowest Rated Genre:")
print(lowest_rated)
````

## **Question 20:** 
### **Between free and paid apps, which type has a higher average rating and more installs?**

````python
# Filter the DataFrame to include only free apps
free_apps = df[df['type'] == 'Free']

# Filter the DataFrame to include only paid apps
paid_apps = df[df['type'] == 'Paid']

# Calculate the average rating for free apps
free_avg_rating = free_apps['rating'].mean()

# Calculate the total number of installs for free apps
free_total_installs = free_apps['installs'].sum()

# Calculate the average rating for paid apps
paid_avg_rating = paid_apps['rating'].mean()

# Calculate the total number of installs for paid apps
paid_total_installs = paid_apps['installs'].sum()

# Print results
print("Free Apps - Average Rating:", free_avg_rating, "| Total Installs:", free_total_installs)
print("Paid Apps - Average Rating:", paid_avg_rating, "| Total Installs:", paid_total_installs)

# Determine which type is better
if free_avg_rating > paid_avg_rating:
    print("Free apps have a higher average rating.")
else:
    print("Paid apps have a higher average rating.")

if free_total_installs > paid_total_installs:
    print("Free apps have more total installs.")
else:
    print("Paid apps have more total installs.")
````
